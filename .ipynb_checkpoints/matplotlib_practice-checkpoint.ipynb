<h3>5.1.2: The Matplotlib Library</h3>

```python
# Import dependencies.
import matplotlib.pyplot as plt
```

<h3>5.1.3: Create Line Charts</h3>

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

<h3>Create a Line Chart Using the Object-Oriented Interface, use ax.plot()</h3>

```python
# Create a Line Chart Using the Object-Oriented Interface, use ax.plot()
# First, create a figure and axes using the object-oriented method
# 1. To create a figure, use fig, ax = plt.subplots().
# 2. To define the axes, use ax.plot().
# 3. Inside the parentheses of ax.plot(), add the axes separated by a comma.

# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# fig, ax = plt.subplots() line of code can also be written like:
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
# will return empty chart
```

```python
# To create a quick graph using the object-oriented interface method
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Note:
# To change figure-level attributes (such as axis labels, a title, or a legend) 
# or save the figure as an image, use fig = plt.figure()
```

```python
# use plt.show(): should only be used once per session
# Create the plot.
# plt.plot(x_axis, y_axis)
# plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

<h3>5.1.4: Modify Line Charts to include line color, marker and label</h3>

```python
# amend the code block to add a new line color, modified line width, and a marker for our data points
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

<h3>5.1.4 SkillDrill: Using the object-oriented approach</h3>

```python
# skilldrill 5.1.4: Using the object-oriented approach
# Change the line color to green, the marker to a diamond, the line width to 2 points.
ax = plt.axes()
ax.plot(x_axis, y_axis, marker="d", color="green", linewidth=2, label='Boston')
# Add a legend for the city of Bostong
ax.legend()
# add a title and axes labels
ax.set_title("PyBer Fare by Month")
# add y-axis limit
ax.set_ylim(0, 45)
# add grid lines.
ax.grid()
```

<h3>5.1.5: Create Bar Charts Using the MATLAB Approach</h3>

```python
# 5.1.5: Create Bar Charts Using the MATLAB Approach
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Vertical bar chart
# Create the plot
plt.bar(x_axis, y_axis)
```

```python
# annotate the bar chart
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Horizontal bar chart
# Create the plot
plt.barh(x_axis, y_axis)
```

```python
# If you want the data on opposite axes, switch the arguments in the barh() function
# Create the plot
plt.barh(y_axis, x_axis)
```

```python
# To invert the y-axis to have the months in ascending order, use the gca() method
# The gca() method means "get current axes."
# Create the plot.
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

<h4>5.1.5 Skilldrill:</h4>

```python
# 5.1.5 Skilldrill:
# Change the bar colors to magenta.
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Invert the y-axis so that January is at the top
plt.gca().invert_yaxis()
```

<h3>Create a Bubble Chart Using the MATLAB Method</h3>

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# adjust the size by multiplying the data in the y-axis by any value
# Let's multiply each data point in the y-axis by 3 and see what happens. 
# To do this, we can iterate through the y-axis data and multiply each data point by 3 and add it to a new list
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
    
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

<h3>5.1.8: Create Pie Charts</h3>

<h4>Create a Pie Chart Using the MATLAB Method</h4>

```python
plt.pie(y_axis, labels=x_axis)
plt.show()
```

```python
# Let's add percentages for each month 
# and "explode" the largest percentage, which is July, the seventh value in the x_axis
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

<h4>Create a Pie Chart Using the Object-Oriented Interface</h4>

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```

<h4>5.1.8 SkillDrill: Using the object-oriented approach, make the following changes to the pie chart</h4>

1.Add a percentage to one decimal place to each wedge of the pie.
2.Increase the figure size to 8x8.
3.Explode the two highest percentage months.
4.Add a shadow.
5.Add a start angle so that January is at the top.
6.Reverse the order so that the month order is in a clockwise direction.
7.Add new colors of your own choosing or use the colors from the previous pie chart.


```python
fig, ax = plt.subplots(figsize=(8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    shadow = True,
    startangle = 90,
    counterclock=False)
plt.show()
```
